```python
import matplotlib.pyplot as plt

# 1. First, ensure no variables are shadowing matplotlib functions
plt.close('all')  # Close any existing figures
del plt  # Delete the plt reference (will re-import below)

# 2. Re-import matplotlib with fresh namespace
import matplotlib.pyplot as plt

# 3. Create your data
x = ['nimra', 'iman', 'kinza', 'shiza']
y = [20, 30, 40, 10]

# 4. Create plot with explicit figure and axes
fig, ax = plt.subplots()  # Modern recommended approach
ax.bar(x, y, color='green', width=0.4)

# 5. Set labels using alternative methods
# Method A - Standard way
ax.set_xlabel('names', fontsize=20)
ax.set_ylabel('age')
ax.set_title('names vs age')

# Method B - Alternative property syntax
# ax.xaxis.label.set_size(20)  # Uncomment if needed

plt.show()
```

```python
speed = [20,30,40,10,20,120,22,50,15]
time  = [1,2,3,2,1,4,5,3,2]
colors = [1,2,3,4,5,6,7,8,9]

sizes = list(map(lambda x: x*40, time))
plt.scatter(speed, time, marker = 'o', c =colors, cmap = 'plasma', edgecolor = 'black',s=sizes )

plt.show()
```

```python
from numpy import random
data = random.randint(15,size = 50)
plt.hist(data, bins = 'auto', linewidth = 2, edgecolor = 'black', color = 'green', rwidth = 1, orientation='horizontal')
plt.show()
```

```python
types = ['food','grocery','bijli','gas']
bills = [10,20,60,10]
plt.pie(bills, explode=[0.0, 0.05, 0.0,0.0], labels = types , autopct = '%0.1f%%', pctdistance = 0.8)
plt.pie([1], colors = 'white', radius = 0.5)
plt.show()
```

```python
# they are used in signal systems
import matplotlib.pyplot as plt 
x=[1,2,3,4,5]
y=[2,-1,4,-3,5]
plt.stem(x,y,linefmt="-.", markerfmt="r+", bottom=0, basefmt="black", label="python")  
# --- linefmt="--" (dashed line)  or ":" (ditted) or "-."(dashed_dot) or "-" (solid) 
            # it can also be linefmt="r--" and so on
# --- markerfmt="r+" yani red color ka marker ho or us ka symbol + ho, or "r*" or 
#               "ro" fot dotted marker, it can also be markerfmt="+"
# --- bottom=1 krny s ab jo horizontal line lgi thi vo 1 pr ho gi show, 
#               bottom=9 kr k dekhna 
# --- basefmt="green" krny s baseline ka color change ho gya
# --- use_line_collection=False
plt.legend()
plt.show()

```

```python
signal1 = (random.randint(low = 20, high = 50, size = 50)).reshape(-1)
signal2 = (random.randint(low = 10, high = 250, size = 50)).reshape(-1)
x= [i for i in range(0,50)]
plt.stem(x,signal1, markerfmt = 'r.', linefmt = 'green', basefmt="black", label = 'lahore' )
plt.stem(x,signal2, markerfmt = '*', label = 'karachi', basefmt = 'black')
plt.legend()
```

```python
datas=[10,20,30,40,50,60,70,200]
plt.boxplot(datas, vert = False , capprops = dict(color = 'green'), patch_artist = True, 
            boxprops = dict(color = 'black'), showmeans = True , whis = 5)
plt.xticks(datas)
plt.show()
```

```python
x=['may','jun','july','aug','sep']
laptops=[2,4,7,1,3]
tablets=[1,1,1,10,1]
l=["laptops","tablets"]
plt.stackplot(x,laptops,tablets, labels=l)
plt.legend()
```
